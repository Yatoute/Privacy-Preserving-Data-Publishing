```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np

# 1) Charger Adult
adult = fetch_ucirepo(id=2)
X = adult.data.features
y = adult.data.targets
print(y.value_counts())
y.loc[:, 'income'] = y['income'].str.replace('.', '', regex=False)
print(y.value_counts())
df = pd.concat([X, y], axis=1).dropna().copy()

# 2) Cibles binaires
df['high_income'] = (df['income'] == '>50K').astype(int)

# 3) Regroupement education -> 3 groupes
low = ['Preschool','1st-4th','5th-6th','7th-8th','9th','10th','11th','12th']
medium = ['HS-grad','Some-college','Assoc-acdm','Assoc-voc']
high = ['Bachelors','Masters','Prof-school','Doctorate']

def group_education(x):
    if x in low: return "Low"
    if x in medium: return "Medium"
    if x in high: return "High"
    return "Other"

df['edu_group'] = df['education'].apply(group_education)

# 4) On verrouille l'ordre des catégories (et on retire 'Other' s'il n'existe pas)
groups = ["Low","Medium","High"]
present = [g for g in groups if (df['edu_group'] == g).any()]
df['edu_group'] = pd.Categorical(df['edu_group'], categories=present, ordered=True)

# 5) Sexe binaire pour la question difficile
df['sex_binary'] = (df['sex'] == 'Male').astype(int)
```

```python
from snsynth.mst import MSTSynthesizer
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance

# 1) Proportions réelles par groupe
prop_real = df.groupby('edu_group')['high_income'].mean().reindex(present)
counts_real = df['edu_group'].value_counts().reindex(present)
display(counts_real)  # tailles des groupes (utile à commenter)
print(prop_real)

# 2) Synthèse MST pour ε=1 et ε=100
features_simple = ['edu_group', 'high_income']
epsilon_values = [1, 100]
synth_props = {}

for eps in epsilon_values:
    synth = MSTSynthesizer(epsilon=eps)
    synth.fit(df[features_simple])
    df_synth = synth.sample(len(df))
    # Harmoniser les catégories
    df_synth['edu_group'] = pd.Categorical(df_synth['edu_group'],
                                           categories=present, ordered=True)
    s = df_synth.groupby('edu_group')['high_income'].mean().reindex(present)
    synth_props[eps] = s.fillna(0.0)

# 3) Écarts numériques
def mae(a, b):
    return float(np.nanmean(np.abs(a - b)))

for eps, s in synth_props.items():
    emd = wasserstein_distance(prop_real.values, s.values)
    print(f"ε={eps} → MAE={mae(prop_real.values, s.values):.4f} | EMD={emd:.4f}")

# 4) Visualisation
x = np.arange(len(present))
w = 0.28
plt.figure(figsize=(9,5))
plt.bar(x - w, prop_real.values, width=w, label='Réel')
plt.bar(x,       synth_props[1].values, width=w, label='Synth ε=1')
plt.bar(x + w,   synth_props[100].values, width=w, label='Synth ε=100')
plt.xticks(x, present)
plt.ylabel("Proportion >50K")
plt.title("Proportion >50K par niveau d'éducation (réel vs synth)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1) Base d'analyse
cols = ['edu_group', 'sex_binary', 'high_income']
df_difficult = df[cols].copy()

# 2) Modèle réel
model_real = smf.logit("high_income ~ C(edu_group) * sex_binary", data=df_difficult).fit()
print(model_real.summary())

# 3) Modèles synthétiques
pred_tables = {}
models_synth = {}

for eps in [1, 100]:
    synth = MSTSynthesizer(epsilon=eps)
    synth.fit(df_difficult[cols])
    df_synth = synth.sample(len(df_difficult))

    # Harmonisation
    df_synth['edu_group'] = pd.Categorical(df_synth['edu_group'],
                                           categories=present, ordered=True)
    df_synth['sex_binary'] = df_synth['sex_binary'].astype(int)
    df_synth['high_income'] = df_synth['high_income'].astype(int)

    model_synth = smf.logit("high_income ~ C(edu_group) * sex_binary", data=df_synth).fit(disp=0)
    models_synth[eps] = model_synth

    # Grille de prédiction (toutes combinaisons edu x sexe)
    grid = pd.DataFrame({
        'edu_group': np.repeat(present, 2),
        'sex_binary': [0,1] * len(present)
    })
    grid['edu_group'] = pd.Categorical(grid['edu_group'], categories=present, ordered=True)
    pred_tables[eps] = model_synth.predict(grid).values.reshape(len(present), 2)

# Prédictions réelles pour la même grille
grid = pd.DataFrame({'edu_group': np.repeat(present, 2),
                     'sex_binary': [0,1] * len(present)})
grid['edu_group'] = pd.Categorical(grid['edu_group'], categories=present, ordered=True)
pred_real = model_real.predict(grid).values.reshape(len(present), 2)

# 4) Indicateur d'écart
def mae_mat(A, B):
    return float(np.nanmean(np.abs(A - B)))

for eps, P in pred_tables.items():
    print(f"ε={eps} → MAE des probabilités (réel vs synth) : {mae_mat(pred_real, P):.4f}")

# 5) Visualisation — Profils par sexe
plt.figure(figsize=(12,5))
for j, sex_label in enumerate(['Femme (0)','Homme (1)']):
    plt.subplot(1,2,j+1)
    x = np.arange(len(present))
    plt.plot(x, pred_real[:, j], marker='o', label='Réel')
    plt.plot(x, pred_tables[1][:, j], marker='o', label='Synth ε=1')
    plt.plot(x, pred_tables[100][:, j], marker='o', label='Synth ε=100')
    plt.xticks(x, present)
    plt.ylim(0, 0.6)
    plt.ylabel("Probabilité prédite (>50K)")
    plt.title(f"Probabilité par niveau d'éducation — {sex_label}")
    plt.legend()

plt.tight_layout()
plt.show()
```
